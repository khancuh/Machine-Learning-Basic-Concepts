```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df
```

# Correlation matrix with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, 
to show a 2D correlation matrix (table) between two discrete dimensions or event types. 
It is very important in Feature Selection

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# JoinPlot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. 
It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

# Univariate Analysis

```python
df.keys()
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pair plot

A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this:
Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Dist plot

Dist plot helps us to check the distribution of the columns feature


```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

    Countplot
    Barplot
    Boxplot
    violinplot

```python
#Countplot
sns.countplot('sex',data=df)
```

```python
#Countplot
sns.countplot(y='sex',data=df)
```

```python
#Bar plot
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(y='sex',x='total_bill',data=df)
```

```python
df.head()
```

```python
#Boxplot
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(y='smoker',x='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories
sns.boxplot(x='day',y='total_bill',hue='smoker',data=df)
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# Violin Plot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot


```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='rainbow')
```
